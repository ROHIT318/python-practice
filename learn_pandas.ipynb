# Grouping and Aggregating

```python
# df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df = pd.read_csv('/content/drive/MyDrive/purchaser.csv')

df.head()
```

```python
df.shape
```

```python
df.columns = ['purchase_date', 'purchaser', 'denomination']
```

```python
df['denomination'].median()
```

```python
# Broad overview or statiscal data. Goes through all columns and provide statistical details of only integer/float columns
df.describe()
```

```python
# To count frequency of an element. If want to count number of elements or rows use count() instead.
df['purchaser'].value_counts()
```

```python
# Get percentage of total values instead
df['purchaser'].value_counts(normalize=True)
```

```python
# splitting of object, apply functions, combine results
# Breaking the dataset into groups by purchasers
purchaser_group = df.groupby(['purchaser'])
purchaser_group.get_group('FUTURE GAMING AND HOTEL SERVICES PR')
# Similar to filter=df['purchaser']=='FUTURE GAMING AND HOTEL SERVICES PR'; df.loc[ filter ];
# Except that group is created for each unique entry of purchaser column
```

```python
# Contains multiple indexes like purchaser and purchase_date
purchaser_group['purchase_date'].value_counts().tail(15)
```

```python
# Accessing data using groupby and than indexing
purchaser_group['purchase_date'].value_counts().loc['YOGA BUILDERS PVT LTD']
```

```python
# 'YOGA BUILDERS PVT LTD' bought 53% of there total bonds on 07/Jul/2022
purchaser_group['purchase_date'].value_counts(normalize=True).loc['YOGA BUILDERS PVT LTD']
```

```python
# Column of grouped data on which you want to apply aggregated function
purchaser_group['denomination'].median()
```

```python
purchaser_group['denomination'].median().loc['ZYDUS HEALTHCARE LIMITED']
```

```python
# Apply multiple aggregate functions in one go
# purchaser_group['denomination'].agg(['median', 'mean'])
purchaser_group['denomination'].agg(['median', 'mean']).loc['A B C INDIA LIMITED']
```

```python
# Another way to apply multiple aggregated functions on a dataframe groupby object
purchaser_group.agg(median_value=('denomination', 'median'), mean_value=('denomination', 'mean'))
```

# Cleaning Data
Casting datatypes and Handling missing values

```python
people = {
    'f_name': ['Rohit', 'Amit', 'Sunny', np.nan, None, 'NA'],
    'l_name': ['Sharma', 'Jain', 'Schafer', np.nan, np.nan, 'Missing'],
    'mail': ['rohit@email.com', 'amit@email.com', 'sunny@email.com', np.nan, None, 'Missing'],
    'age': ['25', '23', '24', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
# Remove null values from dataframe.
# By default below code is like dropna(axis='index', how='any'). Should have an argument inplace=True if want to change on existing dataframe
# axis can have 'index' and 'columns'. how can have 'any' and 'all'
df.dropna()
```

```python
# Deleted all rows as
df.dropna(axis='columns', how='any')
```

```python
df.dropna(axis='index', how='any', subset=['l_name', 'mail'])
```

```python
df
```

```python
# df.replace('NA', np.nan)
# df.replace('None', np.nan)
df.replace('Missing', np.nan)
```

```python
df.isna()
```

```python
df.fillna('NULLS')
```

```python
df.dtypes
```

```python
# type(df['age'][0])
df['age'].dtypes
```

```python
# Change a column data type
df['age'] = df['age'].astype(float)
df['age'].dtypes
```

```python
df['age'].replace([None, 'Missing'], np.nan, inplace=True)
df
```

# Working with dates and time

```python
purchaser_df = pd.read_csv('/content/drive/MyDrive/purchaser.csv')
purchaser_df.head(3)
```

```python
purchaser_df.rename(columns={'Date of Purchase': 'purchase_date', 'Purchaser Name': 'purchaser', 'Denomination': 'denom'},
                    inplace=True)
purchaser_df.head(3)
```

```python
purchaser_df['purchase_date'].head(3)
```

```python
purchaser_df['date'] = pd.to_datetime(purchaser_df['purchase_date'])
purchaser_df['date'].head(3)
```

```python
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
purchaser_df['format_date'] = pd.to_datetime(purchaser_df['purchase_date'], format='%d/%b/%Y')
purchaser_df['format_date'].head(3)
```

**We can convert datetime during reading csv itself.**
- d_parser = lambda x: pd.to_datetime.strptime(x, '%d/%b/%Y') **#Convert a string to date using strptime function**
- pd.read_csv('filename.csv', parse_dates=['date_col1', 'date_col2'], date_parser=d_parser)

```python
purchaser_df.head(3)
```

```python
purchaser_df.iloc[0, 4].day_name()
```

```python
# Get day name of the date
purchaser_df['day_of_week'] = purchaser_df['format_date'].dt.day_name()
purchaser_df.head(3)
```

```python
# Get min and max of the date
print(f'Min date {purchaser_df["format_date"].min()}, Max date {purchaser_df["format_date"].max()}')
```

```python
# Get date difference
purchaser_df["format_date"].max() - purchaser_df["format_date"].min()
```

```python
# Apply filter on date columns, either using entire date or just year only
filter1 = (purchaser_df['format_date'] >= '2023')
filter2 = (purchaser_df['format_date'] <= '2024')

purchaser_df.loc[ filter1 & filter2 ]
```

```python
# Apply filter on date columns, either using entire date or just year only
filter1 = (purchaser_df['format_date'] >= pd.to_datetime('2023-01-01'))
filter2 = (purchaser_df['format_date'] < pd.to_datetime('2025-01-01'))

purchaser_df.loc[ filter1 & filter2 ]
```

```python
# Set date columns as index
purchaser_df.set_index('format_date', inplace=True)
purchaser_df.head(3)
```
