```python
import pandas as pd
```

```python
# df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df = pd.read_csv('/content/drive/MyDrive/purchaser.csv')

df.head()
```

```python
df.shape
```

```python
df.columns = ['purchase_date', 'purchaser', 'denomination']
```

```python
df['denomination'].median()
```

```python
# Broad overview or statiscal data. Goes through all columns and provide statistical details of only integer/float columns
df.describe()
```

```python
# To count frequency of an element. If want to count number of elements or rows use count() instead.
df['purchaser'].value_counts()
```

```python
# Get percentage of total values instead
df['purchaser'].value_counts(normalize=True)
```

```python
# splitting of object, apply functions, combine results
# Breaking the dataset into groups by purchasers
purchaser_group = df.groupby(['purchaser'])
purchaser_group.get_group('FUTURE GAMING AND HOTEL SERVICES PR')
# Similar to filter=df['purchaser']=='FUTURE GAMING AND HOTEL SERVICES PR'; df.loc[ filter ];
# Except that group is created for each unique entry of purchaser column
```

```python
# Contains multiple indexes like purchaser and purchase_date
purchaser_group['purchase_date'].value_counts().tail(15)
```

```python
# Accessing data using groupby and than indexing
purchaser_group['purchase_date'].value_counts().loc['YOGA BUILDERS PVT LTD']
```

```python
# 'YOGA BUILDERS PVT LTD' bought 53% of there total bonds on 07/Jul/2022
purchaser_group['purchase_date'].value_counts(normalize=True).loc['YOGA BUILDERS PVT LTD']
```
