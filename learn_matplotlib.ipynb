**Boilerplate:**
1. `plt.style.use("fivethirtyeight")`: To set a theme for the plots. To check all the avalable one excute `plt.style.available`
2. `plt.legend(loc=location)`: Location can either be a string `ex- ['best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center']` or coordinates in a tuple (x,y) `ex-(0.1,0.1)`
3. `plt.title('Plot Title')`
4. `plt.tight_layout()`
5. `plt.show()`
6. `plt.axvline(value, color=color, label='Line Label', linewidth=2)` / `plt.axhline` : Creates a vertical or horizonatal line on the chart to show the threshold values
7. To change the size of the charts:
```py
from matplotlib import figure
figure(figsize=(4,4), dpi=80)
```
8. To save a visual:
```py
fig, ax = plt.subplots()
fig.savefig('fig_name.png')
```


````python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
import numpy as np

# plt.style.available
plt.style.use("fivethirtyeight")
# figure(figsize=(1,1), dpi=10)
````

# 3. Pie Chart
- Good for comparing 5 to 6 items
- `explode` to move a slice out from the centre
- Other arguments to enhance look of chart:
    - `shadow` gives 3d texture;
    - `startangle` to rotate the chart;
    - `autopct` to label the slices with values
    - `wedgeprops` to create a line between different slices of colors

````python
plt.title("Pie Chart Rainbow")

# data_arr = np.array([30,70])
data_arr = np.random.randint(1,11,size=(6))
legends = np.array(['leg_1', 'leg_2', 'leg_3', 'leg_4', 'leg_5', 'leg_6'])
colors = np.array(['red', 'indigo', 'blue', 'green', 'yellow', 'orange'])
explode = [0,0,0,0,0,0.15]

plt.pie(data_arr, labels=legends, colors=colors,
        explode=explode, startangle=180, shadow=True,
        autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'})
plt.tight_layout()
plt.show()
````

# 4. Stack Plots

**Creating the dataset for visualization**

````python
time_arr = np.arange(1,11,1)
print(f'Time array: {time_arr}')

company_one_score = np.random.randint(80, 90, size=(10))
company_two_score = np.random.randint(90, 100, size=(10))
company_three_score = np.random.randint(70, 80, size=(10))
print(f'Company One Score: {company_one_score}')
print(f'Company Two Score: {company_two_score}')
print(f'Company Three Score: {company_three_score}')
````

````python
labels = ['company_one', 'company_two', 'company_three']
colors = np.array(['green', 'yellow', 'orange'])
plt.stackplot(time_arr, company_one_score, company_two_score, company_three_score,
              labels=labels, colors=colors)
plt.legend(loc='center right')

plt.title('Stack Plot of different Companies')
plt.tight_layout()
plt.show()
````

# 5. Plot Fills

**Creating the dataset for visualization**

````python
time_arr = np.arange(1,11,1)
print(f'Time array: {time_arr}')

# company_one_score = np.random.randint(80, 90, size=(10))
company_one_score = np.array([86, 80, 82, 82, 82, 85, 86, 83, 87, 86])
# company_two_score = np.random.randint(90, 100, size=(10))
company_two_score =  np.array([97, 93, 95, 70, 75, 98, 96, 97, 95, 94])
print(f'Company one score array: {company_one_score}')
print(f'Company two score array: {company_two_score}')
````

````python
plt.plot(time_arr, company_one_score, label='company_one')
plt.plot(time_arr, company_two_score, label='company_two')

plt.fill_between(time_arr, company_one_score, company_two_score,
                 where=(company_one_score > company_two_score),
                 interpolate=True,
                 color='orange', alpha=0.5)

plt.fill_between(time_arr, company_one_score, company_two_score,
                 where=(company_one_score < company_two_score),
                 interpolate=True,
                 color='green', alpha=0.5)

plt.legend(loc='upper left')
plt.title('Plot Fills Example')
plt.tight_layout()
plt.show()
````

# 6. Histograms

````python
age_arr = np.random.randint(18,55,size=(20))
print(f'Age group array: {age_arr}')

figure(figsize=(4,4), dpi=80)
plt.hist(age_arr)

plt.ylabel('Count of people')
plt.xlabel('Age Bins')
plt.title('Age Group Bins')
plt.show()
````

````python
figure(figsize=(4,4), dpi=80)
plt.hist(age_arr, bins=3, edgecolor='black')
plt.xlabel('Age bins')
plt.ylabel('Count of people')
plt.title('Age Group Bins')
plt.show()
````

````python
figure(figsize=(8,4), dpi=80)

bins = np.array([18, 30, 40, 50, 55])

plt.hist(age_arr, bins=bins, edgecolor='black')

vertical_median_value = 32
horizontal_median_value = 3.5
plt.axvline(vertical_median_value, color='red', label='Vertical Median Value', linewidth=2)
plt.axhline(horizontal_median_value, color='green', label='Horizontal Median Value', linewidth=2)

plt.xlabel('Age bins')
plt.ylabel('Count of people')
plt.title('Age Group Bins')
plt.legend()
plt.show()
````

# 7. Scatter Plots

**Data Preparation**

````python
x_arr = np.random.randint(10,100,size=(10))
y_arr = np.random.randint(10,100,size=(10))

print(f'X array: {x_arr}')
print(f'Y array: {y_arr}')
````

**Visualization**

````python
figure(figsize=(4,4), dpi=80)

plt.scatter(x_arr, y_arr, c='orange', marker='X',  # Default marker is circle i.e. o
            edgecolor='blue', linewidth=1, alpha=.9)
plt.xlabel('X co-ordinate')
plt.ylabel('Y co-ordinate')
plt.show()
````

````python
figure(figsize=(4,4), dpi=80)

target_value = np.random.randint(1, 11, size=(10))
print(f'Target Value for color coding: {target_value}')

target_sizes = np.arange(200, 700, 50)
print(f'Target size coding: {target_sizes}')

# c for target color encoding; cmap for colors being related to each other; s changes size of dots
plt.scatter(x_arr, y_arr, s=target_sizes, c=target_value, cmap='Blues')

color_bar_legend = plt.colorbar()
color_bar_legend.set_label('Target Value Color Coding')

plt.xlabel('X co-ordinate')
plt.ylabel('Y co-ordinate')
plt.tight_layout()
plt.show()
````

# 8. Plotting Time Series Data

**Preparing data for visulaization**

````python
dates_x = np.array([datetime(2023, 12, 25), datetime(2023, 12, 26), datetime(2023, 12, 27), datetime(2023, 12, 28),
                  datetime(2023, 12, 29), datetime(2023, 12, 30), datetime(2023, 12, 31)])

data_y = np.random.randint(1, 11, size=(7))
````

**Create Visuals**

````python
figure(figsize=(8,4), dpi=80)
plt.plot_date(dates_x, data_y, linestyle='solid')

plt.gcf().autofmt_xdate() # gcf='get current figure'; autofmt='auto format'

date_format = mpl_dates.DateFormatter('%d-%b-%Y')
plt.gca().xaxis.set_major_formatter(date_format) # gca='det current axis'

plt.xlabel('Dates')
plt.ylabel('Data')
plt.title('Data vs Dates')
plt.tight_layout()
plt.show()
````

# 10. Subplots

**Preparing data for visualization**

````python
plt1_y = np.random.randint(10, 50, size=(10))
plt2_y = np.random.randint(10, 50, size=(10))

plt_x = np.arange(1, 11, 1)

print(f'Plot 1, Y-axis: {plt1_y}')
print(f'Plot 2, Y-axis: {plt2_y}')
print(f'All plot, X-axis: {plt_x}')
````

**Creating visuals:**
- `sharex` flag for sharing X-axis on same column and `sharey` flag for sharing Y-aixs on same row

````python
figure(figsize=(8,4), dpi=80)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True)
# figure(figsize=(16,4), dpi=80)

ax1.plot(plt_x, plt1_y)
ax2.plot(plt_x, plt2_y)

ax1.set_title('Plot 1')
ax1.set_xlabel('Time')
ax1.set_ylabel('Data')
ax1.set_label('Data')

ax2.set_title('Plot 2')
ax2.set_xlabel('Time')

plt.tight_layout()
plt.show()
````
